```python
#load files

file = "population_inflow.csv"
df = pd.read_csv("population_inflow.csv", low_memory=False)
```

```python
#remove uncessarry and redundact data
df_new = df.drop(["CO2", "VAR", "Variable", "GEN", "Gender", "COU", "YEA", "Flag Codes", "Flags"], axis=1)
```

```python
df_new.head()
```

```python
# Rename the "Country" column to "Country of Destination"
df_new ['Country of Destination'] = df_new['Country']
df_new.drop(columns=['Country'], inplace=True)
```

```python
# Pivot the DataFrame to have individual columns for each year
pivoted_df = df_new.pivot_table(index=['Country of birth/nationality', 'Country of Destination'],
                            columns='Year',
                            values='Value',
                            aggfunc='first').reset_index()
```

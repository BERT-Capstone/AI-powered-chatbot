##### 1st

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None, 'display.max_columns', None)
```

```python
df1 = pd.read_csv("API_NY.GDP.MKTP.CD_DS2_en_csv_v2_581470.csv", skiprows=4)
df1.info()
df1.head()
```

```python
# Inspecting for missing values in the dataset
df1.isna().sum()
```

```python
# Inspecting for labelling errors via unique values in each column
cols1 = list(df1.columns)

for col in cols1:
    nUnq = df1[col].nunique()  # Calculate number of unique values for each column
    uValues = df1[col].unique()[:5]  # Retrieve the first 5 unique values for each column
    print([col, nUnq, uValues.tolist() if isinstance(uValues, np.ndarray) else uValues])
```

```python
# Replacing missing values with zero
df1.fillna(0, inplace=True)

# Dropping unnecessary columns
df1.drop (columns=["Indicator Name", "Indicator Code", "Unnamed: 68"], inplace = True)
df1.head()
```

```python
# Save to CSV

df1.to_csv('API_NY.GDP.MKTP.CD_DS2_CLEANED.csv', index=False)
```

##### 2nd

```python
df2 = pd.read_csv("Metadata_Country_API_NY.GDP.MKTP.CD_DS2_en_csv_v2_581470.csv")
df2.info()
df2.head()
```

```python
# Inspecting for missing values in the dataset
df2.isna().sum()
```

```python
# Inspecting for labelling errors via unique values in each column
cols2 = list(df2.columns)

for col in cols2:
    nUnq = df2[col].nunique()  # Calculate number of unique values for each column
    uValues = df2[col].unique()[:10]  # Retrieve the first 10 unique values for each column
    print([col, nUnq, uValues.tolist() if isinstance(uValues, np.ndarray) else uValues])
```

```python
# Replacing missing values
df2['Region'].fillna('Unknown', inplace=True)
df2['IncomeGroup'].fillna('Unknown', inplace=True)

# Dropping unnecessary columns
df2.drop (columns=["SpecialNotes", "Unnamed: 5"], inplace = True)

# Rename 'TableName' column to 'Country Name'
df2.rename(columns={'TableName': 'Country Name'}, inplace=True)

df2.head()
```

```python
# Save to CSV

df2.to_csv('Metadata_Country_API_NY.GDP.MKTP.CD_DS2_CLEANED.csv', index=False)
```
